```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
musicbox_conn = sql.connect("../data/musicbox.sqlite")
musicbox_conn.text_factory = str
```

```python
pd.read_sql("SELECT * FROM sqlite_master", musicbox_conn)
```

## Download data

```python
%%time
pd.read_sql("SELECT COUNT(*) FROM Down", musicbox_conn)
```

```python
pd.read_sql("SELECT COUNT(DISTINCT song_id) FROM Down", musicbox_conn)
```

```python
pd.read_sql("SELECT * FROM Down LIMIT 10", musicbox_conn, parse_dates=['date'])
```

```python
pd.read_sql("SELECT paid_flag, COUNT(*) FROM Down GROUP BY paid_flag", musicbox_conn)
```

```python
%%time
pd.read_sql("SELECT device, COUNT(*) FROM Down GROUP BY device", musicbox_conn)
```
